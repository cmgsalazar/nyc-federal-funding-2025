```python
# initializing a search function

def search_string(keyword):
    """
    This function finds a string in any row in a tabula dataframe.

    Parameter:
    keyword (str): string for keyword or phrase search
    """

    result = df.apply(lambda row: row.astype(str).str.contains(keyword).any(), axis=1).any()
    return result
```
